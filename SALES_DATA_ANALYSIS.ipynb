```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_cleaned = pd.read_csv('/content/cleaned_sales_data.csv')
data_cleaned['Order Date'] = pd.to_datetime(data_cleaned['Order Date'])
data_cleaned['YearMonth'] = data_cleaned['Order Date'].dt.to_period('M')
monthly_sales = data_cleaned.groupby('YearMonth')['Sales'].sum()
monthly_sales.index = monthly_sales.index.astype(str)
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_cleaned = pd.read_csv('/content/cleaned_sales_data.csv')
product_sales = data_cleaned.groupby('Product ID')['Sales'].sum().sort_values(ascending=False)
top_10_products = product_sales.head(10)
print(top_10_products)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_products.index, y=top_10_products.values)
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Product ID')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
